```python
df['col_3']=df['col_1'].apply(lambda x:x*2)
df
#Apply can only work on a single column
#For multiple columns use core python
```

#Merge() -> Used to join the dataframes like in SQL(inner,outer,right,left)

```python
df_1=pd.read_excel('/content/football_worldcup.xlsx')
```

```python
df_2=pd.read_excel('/content/football_worldcup.xlsx')
```

```python
df_1
```

```python
new_score=[[2018,'Russia','France','Croatia',100,64],
            [2022,'Qatar','Argentina','France',100,64]]
new_df=pd.DataFrame(new_score,columns=['Year','Country','Winner','Runners-Up',
                              'GoalsScored','MatchesPlayed'])

df_2 = pd.concat([df_2,new_df], ignore_index=True)
# pd.concat([df2,new_df], ignore_index= True)
df_2
```
