```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings('ignore')
```

```python
sns.get_dataset_names()
```

```python
data = sns.load_dataset('tips')
data.head()
data.dtypes #taking tips and size as our x and y variables for correlation
```

```python
data.isnull().sum()
```

```python
sns.scatterplot(x= data['tip'], y = data['total_bill])
plt.show()
```

```python
correlation = data['tip'].corr(data['total_bill'])
print(round(correlation,2))
```

```python
sns.heatmap(data.corr(numeric_only = True),
                             annot = True,
                             cmap = 'Blues')
plt.show()
```

```python
cor = data.corr(numeric_only = True)
upper = np.triu(cor)
sns.heatmap(cor,
            annot = True,
            cmap = 'Blues',
            mask  = upper,
            vmin = -1, vmax = 1)
plt.show()
# Thus total_bill and tips has a strong positive correlation
```
