```python
import string
import nltk
from nltk.tokenize import sent_tokenize, word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer, SnowballStemmer

nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus ='''Contrary to popular belief!, Lorem Ipsum is not simply random text. \
It has root's in a piece of classical Latin literature from 45 BC,\
making it over 2000 years old!. Richard's McClintock, a Latin professor at \
Hampden-Sydney in Virginia, looked up one of the more obscure Latin words.'''
print(corpus)
```

```python
# converting the corpus to documents
documents = sent_tokenize(corpus)
print(documents)
print()

num = 1
for sentence in documents:
  print(f"Document-{num}:",sentence)
  num+=1
```

```python
for sentence in documents:
  print(word_tokenize(sentence))
```

```python
# printing the list of stopwords
exception = ['no','not','don\t','don']
stop_words = [word for word in stopwords.words('english') if word not in exception]
stop_words.append("'s")
print(stop_words)
```

```python
print(string.punctuation)
```

```python
# removing the stopwords from the documents
for sentence in documents:
  tokenize = word_tokenize(sentence)
  processed_tokens = [word for word in tokenize if word.lower() not in stop_words and word not in string.punctuation]
  print(processed_tokens)
```

```python
for sentence in documents:
  tokenize = word_tokenize(sentence)
  print(tokenize)
```

##Stemming

```python
stem = PorterStemmer().stem
```

```python
word_list = ['programming', 'programmer', 'programs', 'programmed', 'achieve', 'achieving', 'achievement',
             'enjoyed', 'enjoying', 'history', 'historical', 'eating', 'ate', 'orderly', 'elderly', 'friendly']

for word in word_list:
  print(word,'=',stem(word))
```

```python
# issues with words ending with 'li' can be solved with snowballstemmer
snow_ball = SnowballStemmer('english').stem

for word in word_list:
  print(word, '=', snow_ball(word))
```

```python
# stemmming the corpus
documents = sent_tokenize(corpus)
for i in range(len(documents)):
  tokens = word_tokenize(documents[i])
  stemmed_word = [snow_ball(word) for word in tokens if word.lower() not in stop_words and word not in string.punctuation]
  documents[i] = ' '.join(stemmed_word)
print(documents)
```

##Lemmatization

```python
lemmas = WordNetLemmatizer().lemmatize
```

```python
for word in word_list:
  print(word, '=',lemmas(word, pos = 'v'))

# pos --> n-noun, v-verb, a-adjective, r-adverb, s-satellite adjective
```

```python
# lemmatize the corpus
documents = sent_tokenize(corpus)
for i in range(len(documents)):
  tokens = word_tokenize(documents[i])

  lemma_tokens = [lemmas(word,pos = 'v') for word in tokens if word.lower() not in stop_words and word not in string.punctuation]
  documents[i] = ' '.join(lemma_tokens)

print(documents)
```
