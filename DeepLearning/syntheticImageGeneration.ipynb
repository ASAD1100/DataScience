```python
import matplotlib.pyplot as plt
from tensorflow.keras.preprocessing.image import ImageDataGenerator , array_to_img, img_to_array, load_img
```

```python
img = load_img('/content/drive/MyDrive/Colab Notebooks/Projects/DeepLearning/datasets/man.jpeg')
plt.imshow(img)
plt.show()
```

```python
# converting the above image to array
man_imgarr = img_to_array(img)
print(man_imgarr.shape)
```

```python
# reshape the data dimensions to 4 dimensions for use in image generator
man_imgarr = man_imgarr.reshape((1,)+man_imgarr.shape)# the first dim is the batch size i.e., 1 image for processing
print(man_imgarr.shape)
```

```python
datagen = ImageDataGenerator(
    rotation_range = 40,
    width_shift_range = 0.2,
    height_shift_range = 0.2,
    rescale = 1./255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True,
    vertical_flip = True,
    fill_mode = 'nearest'
)
```

```python
# create synthetic images
cnt  = 0
for batch in datagen.flow(man_imgarr, batch_size = 1,
                          save_to_dir = '/content/drive/MyDrive/Colab Notebooks/Projects/DeepLearning/datasets/Image_gen_output',
                            save_prefix = 'man_syn', save_format = 'jpeg'):
  cnt += 1
  if cnt >20:
    break
```
